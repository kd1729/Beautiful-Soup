# Part 1 : Dealing With First Quote

```python
base_url = "https://www.passiton.com/inspirational-quotes?page="
page1 = base_url + "1"
print(page1)
```

```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
res = requests.get(page1)
#print(res.content)
```

```python
soup = BeautifulSoup(res.content)
```

```python
print(type(soup))
```

```python
print(soup.title.text)
```

### Extracting First Quote 

```python
div_quotes = soup.find_all("div", attrs = {"class" : "col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top"})

#total no of quotes on first page
print(len(div_quotes)) 
```

```python
# anchor_tags_inside_a_single_div_quote
x = div_quotes[0].find_all("a")
print("anchor_tags_inside_a_single_div_quote : ", len(x))
```

```python
print(x[0])  # link to the quote iamge
print()
print(x[1])  # link to the tag associsted with that image
```

```python
# Extracting the first quote
print(x[0].img.attrs['alt'].split('#')[0])
```

```python
# Tag of fisrt quote
print(x[1].text)
```

### Extracting image of First Quote

```python
url_first_image = x[0].img.attrs['src']
print(url_first_image)
```

```python
import shutil
```

```python
r = requests.get(url_first_image, stream=True)  # Get request on full_url

with open("First_Image.jpg", 'wb') as f:
    r.raw.decode_content = True
    shutil.copyfileobj(r.raw, f)  # image saved in same directory
    
```

```python
total_pages = 50
```

### displaying all quotes from all pages

### creating a pandas dataframe and importing all data to a csv file
 

```python

page_no_list = []
quote_no_list = []
quote_list = []
tag_list = []
url_image_list = []

for page in range(0, total_pages):
    new_url = base_url + str(page+1)
    res = requests.get(new_url)
    soup = BeautifulSoup(res.content)
    
    div_quotes = soup.find_all("div", attrs = {"class" : "col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top"})
    n = len(div_quotes)
    
    for i in range(0, n):
        page_no_list.append(page+1)
        x = div_quotes[i].find_all("a")
        quote_no_list.append(i+1)
        quote_list.append(x[0].img.attrs['alt'].split('#')[0])
        tag_list.append(x[1].text)
        url_image_list.append(x[0].img.attrs['src'])
        
        
# creating dataframe
import pandas as pd
df = pd.DataFrame({"page_no" : page_no_list, "quote_no" : quote_no_list, "quote" : quote_list, "tag" : tag_list, "iamge_url" : url_image_list})

# saving all quotes in a csv file
df.to_csv("all_quotes.csv", index = False)

print(df.head())
print()
print(df.tail())

```

### saving all images of all pages in their respective directory

```python
main_dir = 'All_Pages_Images'
if not os.path.exists(main_dir):
        os.makedirs(main_dir)

for page in range(0, total_pages):
    new_url = base_url + str(page+1)
    res = requests.get(new_url)
    soup = BeautifulSoup(res.content)
    div_quotes = soup.find_all("div", attrs = {"class" : "col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top"})
    n = len(div_quotes)
    
    inner_dir = os.path.join(main_dir, 'Page_' + str(page+1) + '_Images')
    
    if not os.path.exists(inner_dir):
        os.makedirs(inner_dir)
    
    for i in range(0, n):
        x = div_quotes[i].find_all("a")
        url_image = x[0].img.attrs['src']
        r = requests.get(url_image, stream=True)  
        
        path_name = os.path.join(inner_dir , "quote")        
        image_name = path_name + str(i+1) + ".jpg"
        with open(image_name, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f) 
```
